```python
users = pd.read_csv('https://raw.githubusercontent.com/ben519/DataWrangling/master/Data/users.csv')
sessions = pd.read_csv('https://raw.githubusercontent.com/ben519/DataWrangling/master/Data/sessions.csv')
products = pd.read_csv('https://raw.githubusercontent.com/ben519/DataWrangling/master/Data/products.csv')
transactions = pd.read_csv('https://raw.githubusercontent.com/ben519/DataWrangling/master/Data/transactions.csv')                    
```

```python
print(users.head())
print(sessions.head())
print(products.head())
print(transactions.head())
```

```python
t_and_u = pd.merge(transactions, users, how='left', on='UserID')
```

```python
print(t_and_u[t_and_u['UserID'].isna()])
#print(t_and_u.head(100))
```

```python
inner = pd.merge(transactions, users, how='inner', on='UserID')
print(inner.info())
```

```python
outer = pd.merge(transactions, users, how='outer', on='UserID')
print(outer)
```

```python
print(pd.merge(users, sessions, left_on=['UserID', 'Registered'], right_on=['UserID', 'SessionDate']))
```
