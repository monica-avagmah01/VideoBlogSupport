```python
import matplotlib.pyplot as plt
import random
import numpy as np
import math

%matplotlib inline


Y = np.random.normal(loc=0.5, scale=0.4, size=10000)
Y = Y[(Y > 0) & (Y < 1)]
Y.sort()
X = np.arange(len(Y))


plt.plot(X, Y, color = 'red', alpha = 0.85)
plt.yscale('log')
plt.title('log')
plt.grid(True)
    
#plt.scatter(X, Y, color = 'red', alpha = 0.25)  
#plt.plot(X, Y, color = 'blue', alpha = 0.25)
plt.show()
```
