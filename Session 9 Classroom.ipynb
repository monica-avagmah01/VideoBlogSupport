```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
x = np.linspace(0,5,11)
y = x ** 2
```

```python
plt.plot(x, y, 'r', label = "Example1")
plt.plot(y, x, 'g', label = "Example2")
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.title("Graph example")
plt.scatter(x**2, y**2)
plt.grid(True)
plt.legend()
plt.savefig('1.png')
```

```python
plt.scatter(x, y, c='r', label = "Example1")
plt.scatter(y, x, c='g', label = "Example2")
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.title("Graph example")
#plt.scatter(x**2, y**2)
plt.grid(True)
plt.legend()
plt.savefig('2.png')
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,'r--')
plt.subplot(1,2,2)
plt.plot(x,y,'g*-')
```

```python
plt.subplot(2,2,1)
plt.plot(x,y,'bo-')
plt.subplot(2,2,2)
plt.plot(y,x,'c*--')
plt.subplot(2,2,3)
plt.plot(x**2, y**2, 'g*--')
plt.subplot(2,2,4)
plt.plot(y**2,x**2, 'm+-')
```

## Object oriented features

```python
fig = plt.figure()
axes1 = fig.add_axes([0.1,0.1,0.8,0.8])#main axes
axes2 = fig.add_axes([0.2,0.5,0.4,0.3])#inset axes
axes1.plot(x,y,'b')
axes1.grid()
axes1.set_xlabel('X Label')
axes1.set_ylabel('Y Label')
axes1.set_title('Title')

#Insert figure axes 2
axes2.plot(y,x,'r')
axes2.set_xlabel('X Label2')
axes2.set_xlabel('Y Label2')
axes2.set_title('Title2')
```

```python
fig, axes = plt.subplots()

axes.plot(x,y,'r')

```

```python
fig, axes = plt.subplots(1,2)

i = 0
for ax in axes:
    if (i == 0):
        ax.plot(x,y,'m')
    
    if( i == 1):
        ax.plot(y,x,'c')
    i += 1
    
    
    
```

```python
fig,axes = plt.subplots(figsize=(8,4))

axes.plot(x,y,'y')
axes.grid()
axes.set_xlabel('X Label')
axes.set_ylabel('Y Label')
axes.set_title('Title')
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x, x**2, 'k*-',label = "x**2", alpha = 0.35)
ax.plot(x, x**3, 'yv-', label = "x**3", alpha = 0.75)
ax.legend(loc=0)
```

```python
import seaborn as sns
 
tips = sns.load_dataset('tips')
tips.head()
```

```python
sns.distplot(tips['total_bill'], hist = True, kde = False, bins=30,color='orange')
```

```python
sns.jointplot(x = 'total_bill', y = 'tip', data = tips,kind='reg',color='purple')
```

```python
sns.jointplot(x = 'total_bill', y = 'tip', data = tips,kind='scatter',color='c')
```

```python
sns.jointplot(x = 'total_bill', y = 'tip', data = tips,kind='residual',color = 'm')
```

```python
sns.jointplot(x = 'total_bill', y = 'tip', data = tips,kind='hex',color='pink')
```

```python
sns.pairplot(tips, hue = 'sex', palette = 'coolwarm')
```

```python
sns.pairplot(tips, hue = 'sex', palette = sns.light_palette("navy"))
```

```python
sns.kdeplot(tips['total_bill'])
sns.rugplot(tips['total_bill'])
```

```python
sns.barplot(x='sex',y='tip',data=tips)
```

```python
sns.countplot(x='sex',data=tips)
```

```python
sns.boxplot(x='day',y='total_bill',data=tips,palette = 'rainbow')
```

```python
#This is for kernel density
sns.violinplot(x='day',y='total_bill',data=tips,palette = 'rainbow')
```
