```python
pysqlfn = lambda q:sqldf(q,globals())
```

```python
pysqlfn
```

```python
q = """SELECT
            strftime("%Y", date) as year
            ,SUM(beef) as beef_total
            FROM
                meat
                GROUP BY
                     year
                     LIMIT 5;"""
```

```python
pysqlfn(q)
```
