```python
import matplotlib.pyplot as plt
import random
import numpy as np


X = np.random.uniform(-20, 80, 100)
Y = np.random.normal(0.5, 5, 100)
f_x = 3* X/20 + 5 

data = [0] * 100

for i in range(0, 100):
    data[i] = f_x[i] + Y[i]
    
    
plt.scatter(X, data, color = 'red', alpha = 0.25)  
plt.plot(X, f_x, color = 'blue', alpha = 0.25)
plt.show()
```
